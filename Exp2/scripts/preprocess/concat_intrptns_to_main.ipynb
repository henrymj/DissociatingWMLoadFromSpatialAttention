```python
sub_files = sorted(glob('../DATA-Behavior/*_wInt.csv'))
```

```python
sub_files
```

```python
for f in sub_files:
    main_beh = pd.read_csv(f.replace('_wInt',''))
    main_beh.to_csv(f.replace('_wInt','_Main')) # move to _Main
    
    main_beh.index.name = 'original_trial_index'
    
    int_beh = pd.read_csv(f)
    int_beh.index.name = 'original_trial_index'
    
    full_beh = pd.concat([main_beh, int_beh]).reset_index()
    full_beh.to_csv(f.replace('_wInt',''))
    
```
